# Capacidad de silos de granos

## contexto

Para calcular la capacidad de almacenamiento de un silo cilíndrico en una planta de maíz, el sistema te pedirá ingresar la altura y el radio del contenedor (en metros) para determinar cuántas toneladas puede alberga

## Requerimiento

1. Solicitar el nombre del lote del silo
2. Solicitar el radio(r) del silo en metros (m)
3. Solicitar la altura (h) del silo en metros(m).
4. calcular la capacidad de maiz que se puede almacenar sabiendo que la densidad del maiz es de 0.75 tol/m^3.(tma)

              V=π(3.1416)∗r2∗h

            tma=volumen∗0.75

Prueba: Radio = 3m, Altura = 10m. Salida ≈212.06
 toneladas.

## Explicación del Código:

Aquí se detalla el código en la celda `rKOE9OK1RuN0`:

```python
# ENTRADA: solicitar el nombre del lote del silo
# Solicitar el radio(r) del silo en metros (m)
# Solicitar la altura (h) del silo en metros(m).

nombre = input("Ingrese el nombre del lote del silo: ")
```
*   **`nombre = input("Ingrese el nombre del lote del silo: ")`**: Esta línea solicita al usuario que ingrese el nombre del lote del silo y guarda el texto ingresado en la variable `nombre`.

```python
r = float(input("Ingrese el radio del silo en metros: "))
h = float(input("Ingrese la altura del silo en metros: "))
```
*   **`r = float(input("Ingrese el radio del silo en metros: "))`**: Pide al usuario que ingrese el radio del silo en metros. La función `input()` lee el valor como una cadena de texto, y `float()` lo convierte a un número decimal (punto flotante) para poder realizar cálculos, guardándolo en la variable `r`.
*   **`h = float(input("Ingrese la altura del silo en metros: "))`**: Similar a la anterior, solicita la altura del silo en metros, la convierte a un número decimal y la almacena en la variable `h`.

```python
# PROCESO: Calcular el volumen(V) y la capacidad (tma) del silo

V = (3.1416) * r**2 * h
tma = round( V * 0.75, 3)
```
*   **`V = (3.1416) * r**2 * h`**: Esta línea calcula el volumen `V` del silo cilíndrico usando la fórmula `π * radio² * altura`. Se usa `3.1416` como una aproximación de Pi, `r**2` calcula el radio al cuadrado, y el resultado se multiplica por la altura `h`.
*   **`tma = round( V * 0.75, 3)`**: Calcula la capacidad de maíz en toneladas `tma`. Multiplica el volumen `V` por la densidad del maíz (`0.75 tol/m^3`). La función `round(..., 3)` redondea el resultado a tres decimales.

```python
# SALIDA: informar la capacidad del silo

print("La capacidad del silo", nombre, "es de", tma, "toneladas")
```
*   **`print("La capacidad del silo", nombre, "es de", tma, "toneladas")`**: Esta línea muestra el resultado final al usuario. Imprime una cadena de texto que incluye el nombre del silo (`nombre`) y la capacidad calculada en toneladas (`tma`).

````python
# ENTRADA: solicitar el nombre del lote del silo
# Solicitar el radio(r) del silo en metros (m)
# Solicitar la altura (h) del silo en metros(m).

nombre = input("Ingrese el nombre del lote del silo: ")
r = float(input("Ingrese el radio del silo en metros: "))
h = float(input("Ingrese la altura del silo en metros: "))

# PROCESO: Calcular el volumen(V) y la capacidad (tma) del silo

V = (3.1416) * r**2 * h
tma = round( V * 0.75, 3)

# SALIDA: informar la capacidad del silo

print("La capacidad del silo", nombre, "es de", tma, "toneladas")
````
