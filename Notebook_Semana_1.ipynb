```python
# ENTRDA: Entra ancho y alto de la madera en metros
ancho = input("ingresa el ancho de la venta en metro
1")
ancho = float(ancho)
alto = input("ingresa el alto de la ventana ")
alto = float (alto)
# PROCESO: Calcula la longitud de la madera (lm) en pies
# y el Area del vidrio en metros cuadrados
lm = 2 * (ancho + alto) * 3.25
av = 2 * (ancho * alto)

# SALIDA: sale la longitud de la madera en pies
# el area del vidrio en metros cuadrados
print("Longitud de la madera es ", lm, " pies de longitud")
print("El area del vidrio es ", av, " metros cuadrados")

```
