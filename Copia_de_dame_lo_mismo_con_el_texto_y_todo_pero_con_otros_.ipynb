## 3 Nuevos Ejemplos de Código en Agroindustria

### 1. Balance de Materia (Mezclado)

En la elaboración de jugos o néctares, a veces necesitas ajustar los Grados Brix (azúcar). Este código calcula cuánta agua añadir para diluir un concentrado.

```
# Cálculo de dilución para alcanzar grados Brix deseados
brix_inicial = 65  # Concentrado
brix_objetivo = 12 # Jugo final
masa_concentrado = 100 # kg

# Fórmula: M1 * B1 = M2 * B2
masa_final = (masa_concentrado * brix_inicial) / brix_objetivo
agua_a_añadir = masa_final - masa_concentrado

print(f"Para llegar a {brix_objetivo}° Brix, debes añadir {agua_a_añadir:.2f} kg de agua.")
```

### 2. Clasificación Automática de Calidad (Hortofrutícola)

Imagina un sistema que clasifica fruta por su peso para decidir si va a exportación, mercado nacional o industria (mermeladas).

```
# Clasificación de mango por peso (gramos)
peso_fruto = 450

if peso_fruto > 400:
    categoria = "Exportación Premium"
elif 250 <= peso_fruto <= 400:
    categoria = "Mercado Nacional"
else:
    categoria = "Procesamiento Industrial (Pulpa)"

print(f"El fruto pesa {peso_fruto}g y clasifica como: {categoria}")
```

### 3. Estimación de Vida Útil (Cinética de Deterioro)

Un cálculo rápido para estimar cuántos días le quedan a un producto refrigerado basándose en una tasa de deterioro conocida.

```
# Estimación de días de vida útil restante
calidad_actual = 100  # Escala de 0 a 100
tasa_deterioro_diario = 4.5 # Pierde 4.5% de calidad por día a 5°C
limite_aceptable = 40

dias_restantes = (calidad_actual - limite_aceptable) / tasa_deterioro_diario

print(f"Al producto le quedan aproximadamente {int(dias_restantes)} días antes de perder aptitud comercial.")
```
