# Rendimiento de pulpa en frutas

## Contexto
Calcula el porcentaje de aprovechamiento (R) de tu materia prima ingresando el peso total de la fruta adquirida y el peso de los residuos (cáscaras y semillas) obtenidos durante el procesamiento.

## Requerimiento

1. Solicitar el nombre de la planta.
2. Solicitar el peso total de la fruta 3.comprada en Kg(P_total).
3. Solicitar el peso de los residuos en Kg4(P_resid)
4. calcular el porcentaje de aprovechamiento sasbiendo que:

        Pulpa=P.total−P.resid

        R=(Pulpa/PesoTotal)×100

## prueba del programa

Si se ingresa un peso total de 500kg de fruta comprada y un peso de residuos de 150kg, el programa debee informar como salida 350kg de pulpa osea un rendimiento del 70%.



````python
#ENTRADA: Solicitar el nombre de la planta.
#Solicitar el peso total de la fruta comprada en Kg(P_total).
#Solicitar el peso de los residuos en Kg(P_resid)

nombre = input("Ingrese el nombre de la planta: ")
P_total = float(input("Ingrese el peso total de la fruta comprada en Kg: "))
P_resid = float(input("Ingrese el peso de los residuos en Kg: "))

#PROCESO: calcular el peso de la pulpa(p_pul) y rendimiento R)
p_pul =(P_total - P_resid)
R = (p_pul/P_total)*100

#SALIDA: informar el rendimiento de aprovechamiento
print("La planta", nombre, "tiene un rendimiento del", R, "%","(", p_pul,"Kg)" )
````
