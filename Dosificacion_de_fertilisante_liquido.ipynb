# Dosificacion de fertilisante liquido

## Contexto
ara gestionar la nutrición de un cultivo en una finca, se debe calcular el volumen total de fertilizante necesario para un sistema de riego por goteo. El usuario proporciona la extensión del terreno (hectáreas) y la concentración deseada (litros por cada hectárea) para determinar la cantidad exacta de insumo a aplicar.

## Requerimiento

1. Solicitar el nombre de la finca
2. Solicitar el numero de hectareas totales(H)
3. Solicitar las dosis requeridas (litros/hectarea)(DR)
4. calcular la disis de fertilizante por hectarea sabiendo que el fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua)(FH)

             PT=H∗DR
        DFH=HT∗RelaciondeAgua(500)

Prueba: 5 hectáreas, Dosis 2 L/ha. Salida: 10L de producto y 5000L de agua total.

````python
#ENTRADA:Solicitar el nombre de la finca
#Solicitar el numero de hectareas totales(H)
#Solicitar las dosis requeridas (litros/hectarea)(DR)

nombre = input("Ingrese el nombre de la finca: ")
hectareas = float(input("Ingrese el numero de hectareas totales: "))
dosis = float(input("Ingrese las dosis requeridas (litros/hectarea): "))

#PROCESO:calcular el agua total (AT) y la disis de fertilizante por hectarea (DFH)
AT = hectareas * dosis
DFT = hectareas * 500

#SALIDA: informar la dosis requerida
print("La dosis requerida para la finca", nombre, "es de", AT, "litros de producto y", DFT, "litros de agua total.")
````

El siguiente código calcula la cantidad de fertilizante líquido y agua necesaria basándose en las hectáreas y la dosis requerida:

```python
nombre = input("Ingrese el nombre de la finca: ")
hectareas = float(input("Ingrese el numero de hectareas totales: "))
dosis = float(input("Ingrese las dosis requeridas (litros/hectarea): "))

AT = hectareas * dosis
DFT = hectareas * 500

print("La dosis requerida para la finca", nombre, "es de", AT, "litros de producto y", DFT, "litros de agua total.")
```

### Explicación línea por línea:

1.  `nombre = input("Ingrese el nombre de la finca: ")`:
    *   Esta línea solicita al usuario que ingrese el nombre de la finca y guarda la entrada como una cadena de texto en la variable `nombre`.

2.  `hectareas = float(input("Ingrese el numero de hectareas totales: "))`:
    *   Pide al usuario el número total de hectáreas. La función `input()` lee la entrada como texto, y `float()` la convierte en un número decimal para poder realizar cálculos con ella. El valor se guarda en la variable `hectareas`.

3.  `dosis = float(input("Ingrese las dosis requeridas (litros/hectarea): "))`:
    *   De manera similar, solicita la dosis de fertilizante requerida por hectárea y la convierte a un número decimal, almacenándola en la variable `dosis`.

4.  `AT = hectareas * dosis`:
    *   Calcula el **Agua Total (AT)** de producto de fertilizante necesario. Multiplica el número total de hectáreas por la dosis de fertilizante por hectárea. El resultado se guarda en la variable `AT`.

5.  `DFT = hectareas * 500`:
    *   Calcula la **Dosis de Fertilizante Total (DFT)** de agua necesaria para diluir el fertilizante. Se multiplica el número total de hectáreas por 500, asumiendo una relación de dilución de 1:500 (1 litro de producto por 500 de agua por hectárea, según el contexto). El resultado se guarda en la variable `DFT`.

6.  `print("La dosis requerida para la finca", nombre, "es de", AT, "litros de producto y", DFT, "litros de agua total.")`:
    *   Finalmente, esta línea imprime un mensaje informativo en la consola, mostrando el nombre de la finca, la cantidad total de litros de producto de fertilizante (`AT`) y la cantidad total de litros de agua necesarios (`DFT`).
